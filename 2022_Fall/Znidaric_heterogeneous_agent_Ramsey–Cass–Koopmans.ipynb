# Introduction
Goods: 2 pieces (freetime and a consumer good)
One consumer: 
1. Endowment: $ (L,e) \in \mathbb{R}^2 \rightarrow income $
2. Utility function, u: $ \mathbb{R}^2_+ \rightarrow \mathbb{R}$ 
3. Consumer plan $(f,x) \in [0,L] \times \mathbb{R}_+$

Firm with one production unit(owned by the consumer):
1. Production-function: g: $ \mathbb{R}_+ \rightarrow \mathbb{R}$ 
2. Supply, $y \in \mathbb{R}$
3. Labour demand (hours worked),$ l \in \mathbb{R}_+ $

An equilibrium must be possible: $ x-e=y=g(l), f=L-l $

Walras equilibrium demands equilibrium prices $ w $ and $ p $ and the allocation $(y^*,l^*,f^*,x^*) $ such that 
1. The labour market clears, $ L-f^*=l^*\geq0 $
2. The goods market clears, $y^*=x^*-e\geq0 $
3. Profit-maximization with w og p gives $y^*,l^*,\pi $
Firm profit is of course $\pi $
4. Utility maximization with $w,p$ and $m=w*L+pe+\pi$ gives $f^*$ and $x^*$

We can also use Walras' Law: When the consumers' preferences are monotonous then the goods market is clearing if the labour market clears and vice versa. 

Consumer utility: $ u(f,x)=f^\alpha*x^((\alpha-1))$

The firms' profit-max problem: $\pi(w,p)=max_l pg(l)-wl \rightarrow $ find $l^*(w,p),y^*(w,p),\pi(w,p)$

Clear the labour market (set p as numeraire) and find w with $m=wL+pe+\pi(p,w)$
$$l^*(w,p)=L-f^*(w,p,m) \leftrightarrow y^*(w,p)=x^*(w,p,m)-e$$


Find the equilibrium allocation by inserting in 
$$l^*(w,p),y^*(w,p),f^*(w,p,m),x^*(w,p,m)$$

The firm and consumer problem

Firm: $$\pi(p,w)=max_l pAl^\gamma-wl$$

Labour demand: $l^*(w,p)=\frac{pA\gamma}{w^{(1/(1-\gamma))}}$

Supply of goods: $y^*(w,p)=Al\gamma=A\frac{pA*\gamma}{w}^\frac{\gamma}{1-\gamma}$

Profit: $\pi(w,p)=py-wl=\frac{1-\gamma}{\gamma*w}*\frac{pA\gamma}{w^\frac{1}{1-\gamma}}$

Consumer: $max_f,x f^\alpha*x^1-\alpha$   s.t.  $wf+px=m$
Demand for free time $f^*(w,p,m)=\alpha*m/w$
Demand for goods: $x^*(w,p,m)=(1-\alpha)m/p$


We can also use the 2nd welfare theorem: If consumers' preferences are continous, monotonous and strictly convex, and all firms' prodution-functions strictly concave, then any Pareto-efficient allocation can be implemented as a Walras equilibrium after endowments have been exchanged. 
Note this is sufficient conditions! Not necessary conditions. The absence of a strictly concave production does not exclude that a Pareto-efficient allocation can be implemented as a Walras Equilibrium. 

Example: 

Consumer: $ u(f,x)=f^\frac{1}{2}*x^\frac{1}{2}$ and $L=75$
Firm: $g(l)=20l^2$
Pareto stable allocation: $$l^{PS}=50$$

Can we use the 2nd welfare theorem?
Firm: $g(l)=20*l^2$
$$C(y,w)=wl_b(y)=w\sqrt(y/20)$$
$$MC(y,w)=\frac{w}{2*\sqrt(20)}*y^{\frac{-1}{2}}$$
$\lim_{\to\infty} MC(y,w)=0$
Thus profits go to infinity. Consequence: no prices, p & w such that $y^* =20*l^{PS^{2}}$ will be the firm's Pareto Efficient choice. Why? The production function is convex! (convince yourself it has increasing returns to scale)
What does all of this survive:
More goods, consumers and firms. It survives heterogene and homogenous firms and consumers. 
It wont generally not survive market imperfections [not tested here]
# More Consumers

Market clearing with more consumers $ i \in {1,...,N}$

$$l^*(w,p)= \sum_{i=1}^{N} [L_i-f_i^*(w,p,m)]$$
$$y^*(w,p)= \sum_{i=1}^{N} [x_i^**(w,p,m)-e_i]$$
$$m_i=wL_i+pe_i+\theta_i*\pi(p,w),\theta_i\in(0,1),\sum_{i=1}^{N}\theta_i=1$$
$\theta_i$ is the consumers' share of the profits of the company. Remember we are only assuming 1 company. 

# More firms
Market clearing with more firms $j\in{1,..,J}$
$$\sum_{j=1}^{J}l^*_j(w,p)=L-f^*(w,p,m)$$
$$\sum_{j=1}^{J}y^*_j(w,p)=x^*(w,p,m)-e$$
$$m=wL+pe+\sum_{j=1}^{J}*\pi_j(p,w)$$

Profit-function is 
$$g_j(l)=Bl^\gamma$$
Labour demand is
$$\sum_{j=1}^{J} l_j^*(w,p)=J\frac{pB\gamma}{w}^\frac{\gamma}{1-\gamma}$$

Supply of goods
$$\sum_{j=1}^{J} y_j^*(w,p)=JB\frac{pB\gamma}{w}^\frac{\gamma}{1-\gamma}$$


For multiple consumers and firms every concept is the same but income is 
$$M_i=w\eta_i*L_i+pe_i+\sum_{j=1}^{J}\theta_{ij}*\pi_j$$

Also $\forall j: \theta_{ij}\in(0,1),\sum_{i=1}^{N} \theta_{ij}=1$

# Koopman Economy

```python
# a. Parameters
alpha = 0.4
L = 1
A = 17
gamma = 0.6

# Formula for wage
w = A*gamma*((alpha+gamma*(1-alpha))/(gamma*(1-alpha)*L))**(1-gamma)
print(f'w = {w:.2f}')

# b. Firm
ell = (A*gamma/w)**(1/(1-gamma))
y = A*ell**gamma
pi = (1-gamma)/gamma*w*(A*gamma/w)**(1/(1-gamma))

# c. Income
m = w*L + pi

# d. Consumer
f = alpha*m/w
x = (1-alpha)*m

# e. Figure
ell_vec = np.linspace(1e-8,L-1e-8,10000)
y_vec = A*ell_vec**gamma
u_i_valg = f**alpha*x**(1-alpha)
indiff_vec = (u_i_valg/(L-ell_vec)**alpha)**(1/(1-alpha))

fig = plt.figure(frameon=False,figsize=(6,6/1.5),dpi=100)
ax = fig.add_subplot(1,1,1)

ax.plot(ell_vec,y_vec,label=r'Produktion function: $y = A\ell^{\gamma}$')
ax.plot(ell_vec,pi+w*ell_vec,label=r'Isoprofit/budget line: $y = \pi + w\ell$')
ax.plot(ell_vec,indiff_vec,label=r'Utility: $u(f,x) = u(L-\ell,y)$')
ax.scatter(ell,y,zorder=99,color='black',label=r'Firms optimal choice')
ax.scatter(L-f,x,zorder=99,marker='s',color='black',label=r'Consumers optimal choice')

# Details
ax.set_ylim([0,A*L**gamma])
ax.set_xlim([0,L])
ax.set_xlabel('$\ell$')
ax.set_ylabel('$y$')

fig.tight_layout()
lgd = ax.legend(frameon=True,ncol=1,bbox_to_anchor=(1.05, 1), loc='upper left',)
```

```python
N = 10 # Number of consumers
J = 16 # Number of firms

# Heterogene parameters
alpha_range = [0.1,0.9]
eta_range = [0.5,2.0]
theta_range = [0.0,1.0]
B_range = [2.0,8.0]
gamma_range = [0.1,0.9]
```

```python
# a. Firms
B_hom = A/J**(1-gamma)*np.ones(J)
B_het = np.random.uniform(*B_range,size=(J,))

gamma_hom = np.repeat(gamma,J)
gamma_het = np.random.uniform(*gamma_range,size=(J,))

# b. Preferences
alpha_hom = np.repeat(alpha,N)
alpha_het = np.random.uniform(*alpha_range,size=(N,))

# c. Endowments!
L_hom = L_het = np.repeat(1/N,N) # Time
# e_hom = e_het = np.repeat(0.0,N) # Consumer goods
e_hom = np.repeat(0.0, N)
e_het = np.random.pareto(1.16, size=N) * 70


# Productivity
eta_hom = np.repeat(1.0,N)
eta_het = 1.0 + np.random.uniform(*eta_range,size=(N,))  
    
# Owner ship shares
theta_hom = np.zeros((J,N))
theta_het = np.zeros((J,N))

for j in range(J):
    
    theta_hom[j,:] = np.repeat(1/N,N)
    
    theta_het[j,:] = np.random.uniform(*theta_range,size=N)
    theta_het[j,:] = theta_het[j,:] / np.sum(theta_het[j,:])
    
```

# Function checking for market clearance

```python
def marketclearing(w,J,B,gamma,N,L,e,eta,theta,kun_fejl=True):

    # 1. the firm
    ell = (gamma*B/w)**(1/(1-gamma))
    y = B*ell**gamma
    pi = y - w*ell
    
    # 2. income
    m = w*eta*L + e
    for j in range(J):
        m += theta[j,:]*pi[j]
        
    # 3. the consumer
    f = alpha*m/(eta*w)
    x = (1-alpha)*m
    
    # 4. marketclearing    
    fejl_ell = np.sum(ell) - np.sum(eta*(L-f))
    fejl_y = np.sum(y) - np.sum(x)
    
    # 5. end (written in danish)
    if kun_fejl:
        return fejl_ell
    else:
        print(f'fejl_ell: {fejl_ell:.12f}')
        print(f'fejl_y: {fejl_y:.12f}')
        return ell,y,pi,m,f,x
```

# Homogenous consumers and firms

```python
error = marketclearing(w,J,B_hom,gamma_hom,N,L_hom,e_hom,eta_hom,theta_hom)
print(f'error = {error:.16f}')
```
